```
import random

floors ={1,2,3,4,5}
a_lives_in={1,2,3,4}
b_lives_in={2,3,4,5}
c_lives_in={2,3,4}

check=True

while check:
  b=random.choice(tuple(b_lives_in))
  if b==2:
    c=b+2
    if c in c_lives_in:
      check=False

e=c-3
floors_occ={b,c,e}
a=a_lives_in-floors_occ
floors_occ.update(a)
d=floors-floors_occ
print(a,b,c,d,e)
```
