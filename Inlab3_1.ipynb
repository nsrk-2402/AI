```
#inlab_1
graph = {
    'A' : ['B','C','D'],
    'B' : ['E','F'],
    'C' : [],
    'D' : ['G','H','I'],
    'E' : ['J'],
    'F' : ['K'],
    'G' : ['L'],
    'H' : [],
    'I' : [],
    'J' : [],
    'K' : [],
    'L' : []
}
closed = []
opened = []
def bfs(graph,node):
    opened.append(node)
    
    while opened:
        s = opened.pop(0)
        closed.append(s)
        for neighbour in graph[s]:
            if neighbour not in closed and neighbour not in opened:
                opened.append(neighbour)
    print(closed)

bfs(graph,'A')
```
