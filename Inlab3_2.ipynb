```
#inlab_2
graph = {
    'A' : ['C','B'],
    'B' : ['E','D'],
    'C' : ['G','F'],
    'D' : ['I','H'],
    'E' : ['J'],
    'F' : ['L','K'],
    'G' : ['M'],
    'H' : [],
    'I' : [],
    'J' : [],
    'K' : [],
    'L' : [],
    'M' : []
    
}
closed = []
opened = []

def dfs(graph,node):
    opened.append(node)
    while opened:
        n = opened.pop()
        closed.append(n)
        for neighbour in graph[n]:
            if neighbour not in closed and neighbour not in opened:
                opened.append(neighbour)
    print(closed)

dfs(graph,'A')
```
